```python
data=data[data.duplicated()]
data
```

```python
data=data.drop_duplicates()
data
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler,LabelEncoder

# Select only numerical features for scaling
numerical_features = data.select_dtypes(include=['float64', 'int64']).columns

# Standard Scaling
standard_scaler = StandardScaler()
data_standard_scaled = data.copy()
data_standard_scaled[numerical_features] = standard_scaler.fit_transform(data[numerical_features])

# Display the first few rows of the Standard Scaled dataframe
print("\nStandard Scaled DataFrame:")
print(data_standard_scaled.head())
```

```python
# Min-Max Scaling
min_max_scaler = MinMaxScaler()
data_min_max_scaled = data.copy()
data_min_max_scaled[numerical_features] = min_max_scaler.fit_transform(data[numerical_features])

# Display the first few rows of the Min-Max Scaled dataframe
print("\nMin-Max Scaled DataFrame:")
print(data_min_max_scaled.head())
```

# Encoding Techniques

```python
import pandas as pd
categorical_columns = data.select_dtypes(include=['object']).columns
# Apply One-Hot Encoding to categorical variables with less than 5 unique categories
for col in categorical_columns:
    if data[col].nunique() < 5:
        data= pd.get_dummies(data, columns=[col], prefix=[col], drop_first=True)
# Display the first few rows of the dataframe after One-Hot Encoding
print("\nDataFrame after One-Hot Encoding:")
print(data.head())
```

```python
categories_to_encode = data.select_dtypes(include=['object']).columns  # Select categorical columns
for col in categories_to_encode:
    if data[col].nunique() >5:# Check number of unique categories
        label_encoder = LabelEncoder()
        data[col] = label_encoder.fit_transform(data[col])

# Display the DataFrame after Label Encoding
print("\nDataFrame after Label Encoding:")
print(data)
```

# Feature Selection

```python
from sklearn.ensemble import IsolationForest
# Identify outliers using Isolation Forest
clf = IsolationForest(contamination=0.05, random_state=42)
outlier_preds = clf.fit_predict(data.dropna())
# Create a mask to identify outliers
outlier_mask = outlier_preds == -1
# Remove outliers from the dataframe
data_no_outliers = data.loc[~outlier_mask]
# Display the dataframe after removing outliers
print("\nDataFrame after removing outliers:")
data_no_outliers
```
